```python
# import py_entitymatching package
import py_entitymatching as em
# import py_stringmatching package
import py_stringmatching as sm
```

```python
# specify filepaths for tables A and B. 
path_A = '/Users/Hacker/Documents/Courses/Fall 2016/CS638/projects/CS638/stage3/formattedTableA.csv' # path_A is the file path where table_A.csv is stored
path_B = '/Users/Hacker/Documents/Courses/Fall 2016/CS638/projects/CS638/stage3/formattedTableB.csv' # path_B is the file path where table_B.csv is stored
```

```python
# read table A; table A has 'ID' as the key attribute
A = em.read_csv_metadata(path_A, key='id')

# read table B; table B has 'ID' as the key attribute
B = em.read_csv_metadata(path_B, key='id')
```

```python
# Perform Attribute equivalence blocking: block A, B  using pages
ab = em.AttrEquivalenceBlocker()
Cpages = ab.block_tables(A, B, 'pages', 'pages',
                    l_output_attrs=['title', 'authors', 'publisher', 'ISBN13'], 
                    r_output_attrs=['title', 'authors', 'publisher', 'ISBN13'])
```

```python
# Perform Attribute equivalence blocking on publishedYear, publishedMonth, publishedDay
# Essentially multi-attributes equivalence blocking in BlackBoxBlocker
bb = em.BlackBoxBlocker()
```

```python
# Detect the inconsistent type issue of "publishedYear", perform the corresponding type conversion so er
A['publishedYear'] = A['publishedYear'].convert_objects(convert_numeric=True)
```

```python
# Define the black_box function used to check the equivalance of date between tuples
def date_equivalence_block(x, y):
    x_day = x['publishedDay']
    y_day = y['publishedDay']
    x_month = x['publishedMonth']
    y_month = y['publishedMonth']
    x_year = x['publishedYear']
    y_year = y['publishedYear']
    if x_day != y_day or x_month != y_month or x_year != y_year:
        return True
    else:
        return False
```

```python
# Pass the date_equivalence_block to black box function
bb.set_black_box_function(date_equivalence_block)
Cdate = bb.block_tables(A, B, l_output_attrs=['title', 'authors', 'publisher', 'ISBN13'], r_output_attrs=['title', 'authors', 'publisher', 'ISBN13'], l_output_prefix='ltable_', r_output_prefix='rtable_', verbose=False, show_progress=True, n_jobs=4)
```

```python
# Checking the output of the date blocker
Cdate
```

```python
# Union the  to get the table C
C = em.combine_blocker_outputs_via_union([Cpages,Cdate])
```

```python
# Use the debugger to check whether the current blockers filtered too many tuple pairs that will actually match
# Changing the output_size to see how many "matched tuple pairs" have been filtered
dbg = em.debug_blocker(C, A, B, output_size=1000)
```

```python
def isbn_partial_block(x,y):
    try:
        isbn_A = str(int(x['ISBN13']))
    except Exception:
        isbn_A = "0000000000000"
  
    isbn3_head_A = isbn_A[0:3]
    isbn3_tail_A = isbn_A[10:]
    
    try:
        isbn_B = str(int(y['ISBN13']))
    except Exception:
        isbn_B = "0000000000000"
    
    isbn3_head_B = isbn_B[0:5]
    isbn3_tail_B = isbn_B[8:]
    
    if(isbn3_head_A != isbn3_head_B or isbn3_tail_A != isbn3_tail_B):
        return True
    else:
        return False
    
```

```python
bb.set_black_box_function(isbn_partial_block)
Cisbn = bb.block_tables(A, B, l_output_attrs=['title', 'authors', 'publisher', 'ISBN13'], r_output_attrs=['title', 'authors', 'publisher', 'ISBN13'], l_output_prefix='ltable_', r_output_prefix='rtable_', verbose=True, show_progress=True, n_jobs=4)
```

```python
Cisbn
```

```python
C = em.combine_blocker_outputs_via_union([Cpages, Cdate, Cisbn])
```

```python
dbg = em.debug_blocker(C, A, B, output_size=1000)
```
